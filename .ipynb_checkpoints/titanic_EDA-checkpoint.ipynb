```python
df=df.drop('Cabin',axis=1)
```

```python
df.head()
```

```python
df["Age"].dtype
```

```python
sns.boxplot(df['Age'])
```

```python
np.nanmedian(df['Age'])
```

```python
df['Age']=df['Age'].fillna(np.nanmedian(df['Age']))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Embarked'],inplace=True)
```

```python
sns.pairplot(df,hue='Survived')
```

```python
sns.countplot(x='Survived',data=df, hue='Sex')
```

```python
sns.countplot(x='Survived',data=df, hue='Pclass')
```

```python
sns.lmplot(x='Age',y='Fare',data=df,hue="Survived")
```

```python
sns.jointplot(x='SibSp',y="Parch",data=df,kind='scatter',hue='Survived')
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```
